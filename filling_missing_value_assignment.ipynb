```python
import pandas as pd
```

```python
df=pd.read_csv('framingham.csv')
df.head()
```

```python
df.rename(columns={'male':'gender'},inplace=True)
```

```python
df.isnull().sum()
```

```python
def missing_value(column_name):

    if df.dtypes[column_name]=='object':
     replace_value=df[column_name].mode()[0]
    
    elif df[column_name].std() >  df[column_name].mean():
     replace_value=df[column_name].median()
    
    else:
     replace_value=df[column_name].mean()
    df[column_name].fillna(replace_value, inplace=True)
    return df

missing_value('education')
missing_value('cigsPerDay')
missing_value('BPMeds')
missing_value('totChol')
missing_value('BMI')
missing_value('heartRate')
missing_value('glucose')


```
