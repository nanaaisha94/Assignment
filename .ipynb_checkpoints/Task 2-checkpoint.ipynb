```python
import pandas as pd
```

### % of male and female

```python
round(Diversity_Inclusion['Gender'].value_counts(normalize=True)*100)
```

### no of men and women

```python
Diversity_Inclusion['Gender'].value_counts()
```

### no of leavers

```python
Diversity_Inclusion['FY20 leaver?'].value_counts()
```

```python
print('number of FY20 leavers=47')
```

### % of employees promoted in FY21

```python
round(Diversity_Inclusion['Promotion in FY21?'].value_counts(normalize=True)*100,1)
```

```python
print('percentage of employees promoted in FY21 = 10.2%')
```

```python
round(Diversity_Inclusion.groupby('Gender')['Promotion in FY21?'].value_counts(normalize=True)*100,1)
```

```python
print('males promoted = 11.2%')
print('females promoted = 8.8%')
```

### % of hired male and female employees

```python
round(Diversity_Inclusion.groupby('Gender')['New hire FY20?'].value_counts(normalize=True)*100,1)
```

```python
print('male hired = 10.8%')
print('female hired = 16.6%')
```

### job level after FY20 promotion

```python
Diversity_Inclusion.groupby('Job Level after FY20 promotions')['Gender'].value_counts()
```

### Average performance ratings (FY20)

```python
round(Diversity_Inclusion.groupby('Gender')['FY20 Performance Rating'].mean())
```

### Average performance ratings (FY19)

```python
round(Diversity_Inclusion.groupby('Gender')['FY19 Performance Rating'].mean())
```

### no of employees by region

```python
Diversity_Inclusion.groupby('Gender')['Region group: nationality 1'].value_counts()
```

### Average age of employee

```python
round(Diversity_Inclusion['Age @01.07.2020'].mean())
```

### Age group by job level

```python
Diversity_Inclusion.groupby('Age group')['Job Level after FY20 promotions'].value_counts()
```
